```python
import requests
import urllib3

urllib3.disable_warnings()
URL = 'https://www.youtube.com/search'

pattern = input("Enter String to search for: ")
PARAMS = {'q' : pattern}
req = requests.get(url = URL, params = PARAMS , verify = False)
print(req.status_code)
print('-'*100)
print(req.is_redirect)
print('-'*100)
print(req.headers)
print('-'*100)
print(req.text)
print('-'*100)
print(req.url)
```
