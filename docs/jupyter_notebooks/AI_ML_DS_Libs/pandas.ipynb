##### Merging DataFrames

```python
cars = pd.DataFrame({ 'Car' : [ 'Chevrolet Volt ', 'Cadillac CTS ', 'Jaguar F Type ', 'Hyundai Sonata ', 
                               'Ford Fiesta ', 'Tesla Model S ', 'Volkswagen GTI ', 'Volvo S6 ' ], 
                      'Colors' : [ 'Red ', 'Silver ', 'Yellow ', 'Black ', 'Silver ', 'Red ', 
                                   'Blue', 'Black ' ], 
                      'CarID' : [ 21 , 34 , 11 , 23 , 17 , 39 , 9 , 12 ] })

cars
```

```python
more_cars = pd.DataFrame({ 'Car' : [ 'Audi S6', 'Porsche Panamera' ], 
                           'Colors' : [ 'Red', 'Silver' ], 
                           'CarID' : [ 14, 5 ] })

more_cars
```

```python
pd.concat([cars, more_cars])
```

```python
pd.concat([cars, more_cars], ignore_index=True)
```

```python
cars_orders = pd.DataFrame({ 'OrderID' : [201, 202, 203, 204, 205, 206, 207, 208], 
                             'Quantity' : [1, 2, 1, 8, 5, 10, 3, 1], 
                             'DownPayment' : [ '$5,959', '$19,500', '$33,102', '$5,700', '$9,201', '$78,950', '$21,310', '$11,321' ], 
                             'CarID' : [23, 39, 17, 9, 12, 21, 34, 11] })

cars_orders
```

```python
pd.concat([cars, cars_orders])
```

```python
pd.concat([cars, cars_orders], axis=1)
```

```python
pd.merge(cars, cars_orders)
```

```python
pd.merge(cars, cars_orders, on="CarID")
```

```python
extended_cars = pd.concat([cars, more_cars], ignore_index=True)
pd.merge(extended_cars, cars_orders, on="CarID")
```

```python
pd.merge(extended_cars, cars_orders, on="CarID", how="left")
```

```python
pd.merge(extended_cars, cars_orders, on="CarID", how="right")
```
